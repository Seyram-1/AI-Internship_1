## **Calculate molecular descriptors**

###  **LogP, MW and RB**

```python
import numpy as np
```

```python
from rdkit.Chem import Descriptors
```

```python
def generate(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem)
        moldata.append(mol)

    baseData= np.arange(1,1)
    i=0
    for mol in moldata:

        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_MolWt = Descriptors.MolWt(mol)
        desc_NumRotatableBonds = Descriptors.NumRotatableBonds(mol)

        row = np.array([desc_MolLogP,
                        desc_MolWt,
                        desc_NumRotatableBonds])

        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1

    columnNames=["MolLogP","MolWt","NumRotatableBonds"]
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)

    return descriptors
```

```python
df = generate(dataset['PUBCHEM_EXT_DATASOURCE_SMILES'])
df
```

#   **Aromatic proportion**

### Number of aromatic atoms

```python
#Computing for a single molecule.
p = Chem.MolFromSmiles(dataset['PUBCHEM_EXT_DATASOURCE_SMILES'][0])
```

```python
aromatic_atoms = [p.GetAtomWithIdx(i).GetIsAromatic() for i in range(p.GetNumAtoms())]
aromatic_atoms
```

```python
def AromaticAtoms(p):
  aromatic_atoms = [p.GetAtomWithIdx(i).GetIsAromatic() for i in range(p.GetNumAtoms())]
  aa_count = []
  for i in aromatic_atoms:
    if i==True:
      aa_count.append(1)
  sum_aa_count = sum(aa_count)
  return sum_aa_count
```

```python
AromaticAtoms(p)
```
